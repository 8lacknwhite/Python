Aim: Write a python program to implement List operations (Nested List, Length, Concatenation, Membership, Iteration, Indexing and Slicing

**Algorithm**:


1.   Start Program
2.   Inialtize a List
3.   To find length of the list use len function
4.   To concatinate 2 list use the + operator
5.   To check if an item is in the list then use in and for keywords.
6.   For indexing give 2 parameters
7.   For slicing give a ratio in the parameter.
8.   End Program



```python

nested_list = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

length = len(nested_list)
print("Length of nested list:", length)

concat = nested_list + [[12, 13, 14]]
print("Concatenated list:", concat)

member = 5 in [item for sublist in nested_list for item in sublist]
print("Is 5 in nested list:", member)

for sublist in nested_list:
    for item in sublist:
        print(item)

index = nested_list[1][2]
print("Indexing:", index)

slicing = nested_list[1:3]
print("Slicing:", slicing)
```

(b) Write a python program to implement List methods (Add, Append, Extend & Delete)

**Algorithm**:


1.   Start Program
2.   Initiate a list
3.   To add an item to the list use either the + operator or use the append keyword
4.   To add multiple items use the extend keyword
5.   To delete an item use either the remove or del keyword.
6.   End program.



```python

numbers = [1, 2, 3]

numbers = numbers + [4]
print("Added using '+' operator:", numbers)

numbers.append(5)
print("Added using 'append' method:", numbers)

more_numbers = [6, 7, 8]
numbers.extend(more_numbers)
print("Extended list:", numbers)

numbers.remove(5)
print("Deleted using 'remove' method:", numbers)

del numbers[1]
print("Deleted using 'del' statement:", numbers)
```
