**Algorithm**:-

1.   Start Program
2.   Import collections
3.   Define a function for bfs traversal with 2 variables graph, root
4.   Define 2 more variable visited, queue as a set and use the dque function of collection to deque root each time from the queue
5.   Start a while loop with condition being queue and pop the last previous element from quque and print the poped element
6.   Start a for loop inside the while loop to check if the vertex we came to visit is already visited or not
7.   If it is not visited then append that vertex into the queue
8.   Input the graph and state the root
9.   Call the bfs function
10.  End Program



```python
import collections

def bfs(graph,root):
  visited, queue = set(), collections.deque([root])
  visited.add(root)
  while queue:
    vertex= queue.popleft()
    print(str(vertex)+" ", end="")

    for neighbour in graph[vertex]:
      if neighbour not in visited:
        visited.add(neighbour)
        queue.append(neighbour)

if __name__ == '__main__':
  graph={0: [1, 2], 1: [2], 2: [3], 3: [1, 2,0]}
  bfs(graph,1)
```
