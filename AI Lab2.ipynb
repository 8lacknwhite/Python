**Aim**: To implement a ChatBot using python

**Algorithm**: 


1.   Start program
2.   Define a function chatbot()
3.   Intiate a flag variable
4.   Use a while loop to enter all the commands




```python
def chatbot():
  print ("Hi I am a chatbot, How may I assist you today")
  flag= True
  while flag == True:
    user_input= input("May I ask your name: ")
    print("Hi", user_input,"How is your day going")
    user_answer= input("Answer in Good or bad: ")
    if user_answer.lower() in ["good"]:
      print("That is great to know, How may I help you to make your day even better")
    elif user_answer.lower() in ["bad"]:
      print("No worries let me help your day get better")
    user_end= input()
    if user_end.lower() in ["bye","exit","quit"]:
      print("Have a nice day")
    else:
      print("Sorry i didnt understand you")
    flag = False
    
chatbot()
```
