A) **Aim:-** Python Program to Print Multiplication Table of a Number.

**Algorithm**


1.   Start the Program
2.   Take Int input from the user
3.   Use a For loop to print (x*1) until the end of loop execution.

```python
x=int(input("Enter a number= "))
for i in range(1,11):
    y = x*i
    print(x,"*",i,"=",y)
```

B) **Aim-** To check whether the number is prime or not

**Algorithm**


1.   Start program
2.   Take input from the user and declare a bool variable flag.
3.   Make an if else ladder for the input 1 and input greater than 1.
4.   When input is 1 print not a prime number.
5.   When input is greater than 1 then use a for loop to check whether the remainder is 0 or not.
6.   If a remainder is 0 then set flag to true and break out of the loop.
7.   If flag is false then it prime, if flag is true then it is not prime.
8.   End program.


```python
x = int(input("Enter a Number= "))
flag= False
if x==1:
  print(x,"is not a prime numebr")
elif x>1:
  for i in range(2,x):
    if (x%i)==0:
      flag= True
      break
if flag:
  print(x,"is not a prime number")
else:
  print(x,"is a prime number")
```

C) **Aim-** Program to print factorial of a number

**Algorithm**


1.   Start Program
2.   Take input from user and declare an int variable and set value to 1.
3.   Use an If and else ladder for input less than 0, equal to 0 and greater than 0
4.   If input less than 0 then print factorial doesnt exist.
5.   If input is equal to 0 then print factorial is 1.
6.   If input is greater than to 0 then use a for loop to multiply the int varibale with the index till the for loop excution.
7.   Print the int variable.
8.   End the Program



```python
x = int(input("Enter a number= "))
y=1
if x < 0:
   print("factorial does not exist")
elif x == 0:
   print("The factorial of 0 is 1")
else:
   for i in range(1,x+ 1):
       y = y*i
   print("The factorial of",x,"is",y)
```
