**Aim**:- To write a python program to remove punctuations from a given string

**Algorithm**:-

1.   Start program
2.   Define a list of all punctuations
3.   Get the Input string
4.   Define another variable for storing the punctuationless sentence
5.   Run a for loop to check if the charecter is a punctuation or not
6.   If it is then skip that and if not then add it into the no punctuation list
7.   Print the no punctuation list
8.   End program



```python
punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''

string= "!! HI ., Im {} nimit"

no_punct=""

for char in string:
  if char not in punctuations:
    no_punct=no_punct + char

print(no_punct)
```

**Aim**:-Write a python program to sort sentence in alphabetical order

**Algorithm**:-

1.   Start Program
2.   Take input string
3.   Make a list containing each word of the input string using .split() function
4.   Sort the list
5.   Print the list
5.   End Program



```python
my_str = input("Enter a string: ")

words = [word.lower() for word in my_str.split()]

words.sort()
for word in words:
  print(word)
```
