**Aim**: Write a python program to generate Calendar for the given month and year

**Algorithm**:


1.   Start program
2.   Import calender library
3.   Define a function with 2 parameters
4.   use monthcalender function of the calender library
5.   Initiate a for loop for printing the calender.
6.   Take month and year input from the user and call the function.
6.   End Program



```python
import calendar

def generate_calendar(year, month):
    
    cal = calendar.monthcalendar(year, month)


    print(" Mo Tu We Th Fr Sa Su")
    for week in cal:
        for day in week:
            if day == 0:
                print("   ", end="")
            else:
                print(" %2d" % day, end=" ")
        print()


year = int(input("Enter the year: "))
month = int(input("Enter the month (1-12): "))


generate_calendar(year, month)
```

**Aim**: Write a python program to implement Simple Calculator program.

**Algorithm**:


1.   Start program
2.   Define 4 functions for addition , subtraction, multiplication and division respectively each having 2 parameters.
3.   For addition use + operator, subtraction - , multiplication * and division /
4.   Take input what operation user wants using choice keyword
5.   Take 2 numbers as input for the operations
6.   



```python
def add(num1, num2):
    return num1 + num2

def subtract(num1, num2):
    return num1 - num2

def multiply(num1, num2):
    return num1 * num2

def divide(num1, num2):
    return num1 / num2

print("Select operation.")
print("1. Add")
print("2. Subtract")
print("3. Multiply")
print("4. Divide")

choice = int(input("Enter choice(1/2/3/4): "))

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

if choice == 1:
    print(num1, "+", num2, "=", add(num1, num2))

elif choice == 2:
    print(num1, "-", num2, "=", subtract(num1, num2))

elif choice == 3:
    print(num1, "*", num2, "=", multiply(num1, num2))

elif choice == 4:
    print(num1, "/", num2, "=", divide(num1, num2))
else:
    print("Invalid input")
```
